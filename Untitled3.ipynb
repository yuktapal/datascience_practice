CHANGE THE BRIGHTNESS OF IMAGE

```python
def change_bright_bgr(image,add_val):

  imge_hsv=cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
  hue,sat,val=cv2.split(imge_hsv)
  new_val=val.astype(np.int64)+add
  new_val[new_val>255]=255
  new_val_uint8=new_val.astype(np.uint8)
  imge_new_hsv=cv2.merge([hue,sat,new_val_uint8])
  imge_new_bgr=cv2.cvtColor(imge_new_hsv,cv2.COLOR_HSV2BGR)
  return imge_new_bgr
```

```python
imge_new_hsv=cv2.merge([hue,sat,new_val_uint8])
```
